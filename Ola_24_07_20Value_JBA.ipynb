```python
with open("ola.pre", "r") as fname:
    v = []
    g_ref  = []
    fname.seek(249)
    contents = fname.readlines()
    #print(contents[0:10])
    for line in contents: # loops through the contents of the file and returns the Grid-ref and values
        if line.startswith("Grid-ref="): # this removes the string and allows access to the x and y refs
            nline = line.strip('Grid-ref=') # strips and makes lists line by line
            nline = nline.rstrip() #gets rid of newline 
            nline = nline.lstrip() # gets rid of white space 
            nline = nline.split(',') # 
            nline =  [nline]*12*10
            
            for grid in nline:
                gref = g_ref.append(grid)
    for row in contents: #loops through the contents of the file and returns the Grid-ref and value
        if not row.startswith("Grid-ref="): # strips grid references and allows access to value
            nnrow  = row.strip('\n')
            #print(nnrow)
            nrows = nnrow.split() #split the value rows and columns and make lists of them
    
            v.append(nrows) #adds items in nrows to an empty list value 
    #print(g_ref)
    pd_ref = pd.DataFrame(g_ref, columns=['Xref', 'Yref'])

                
```
